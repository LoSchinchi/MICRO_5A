```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size_x, size_y = 100, 100
image = np.zeros((size_x, size_y), dtype='int16')
```

```python
plt.imshow(image)
```

```python
image[0:5,:] = 255
plt.imshow(image)
```

```python
image[-5:,:] = 255
plt.imshow(image)
```

```python
image[:,0:5] = 255
image[:,-5:] = 255
plt.imshow(image)
```

```python
image[size_x // 2, size_y // 2] = 100
plt.imshow(image)
```

```python
image[size_x // 2 - 5: size_x // 2 + 5, size_y // 2 - 5: size_y // 2 + 5] = 100
plt.imshow(image)
```

```python
size_x, size_y = 256, 256

img2 = np.zeros((size_x, size_y), dtype='int16')
img2[:] = np.arange(0, size_y)
plt.imshow(img2, cmap='hsv')
```

```python
img2[size_x // 2:] = 0
plt.imshow(img2, cmap='hsv')
```

```python
img3 = np.zeros((size_x, size_y), dtype='int16')
n_dots = 10000
x = np.random.randint(0, size_x, n_dots)
y = np.random.randint(0, size_y, n_dots)
```

```python
img3[x,y] = 255
plt.imshow(img3, cmap='magma')
```
